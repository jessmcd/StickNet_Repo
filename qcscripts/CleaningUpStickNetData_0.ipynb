```python
import os
import linecache
import glob
import numpy as np
import pandas as pd
import datetime as dt


#######################################################
# USER EDIT HERE =>

directory = '/Users/jessmcd/Documents/MyPassport_backup/Meso18_19/RawData/December_datadump/raw_test/'

#######################################################
```

```python
# trying to find an automatic way to fix messed up file names using headers? be CAREFUL as this can be dangerous
def month_guess(day):
    ''' edit based on given info in file'''
    if day < 18:
        month=12
    if day >= 20:
        month=11
    return month

year = 2018


for f in files:

    fstr = f.split('/')[-1]
    try:
        time =dt.datetime.strptime(fstr[-19:-4], '%Y%m%d_%H%M%S')

    except: # wrong format
        filepath = directory+fstr
        header_text = linecache.getline(filepath,2)
        linecache.clearcache()

        # get date
        datestr = header_text.split(',')[0]
        day = int(datestr[0:2])
        month = month_guess(day)
        time = dt.datetime(year, month, day, int(datestr[3:5]), int(datestr[5:7]), int(datestr[7:9]))

        # ID 
        ID = fstr[0:5]

        # build new name
        newname = '{}{}_{}.txt'.format(directory, ID, time.strftime('%Y%m%d_%H%M%S'))

        print('old name', fstr)
        print('new name', newname.split('/')[-1])

        os.rename(f, newname)  
```
