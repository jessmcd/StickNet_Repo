```python
import os
import linecache
import glob
import numpy as np
import pandas as pd
import datetime as dt
%pylab inline

import sys
vse_folder = '/Users/jessmcd/Documents/GitHub_repos/vortexse/'
sys.path.insert(0, vse_folder)
from probe_info import probe_locs_2019 as probe_locs
```

```python
fpath = '/Users/jessmcd/Documents/MyPassport_backup/Meso18_19/dailyfiles/daily_files_new_QC/'


for probe in probe_locs.keys():
    print(probe)
    
    datadir = f'/Users/jessmcd/Documents/MyPassport_backup/Meso18_19/dailyfiles/daily_files/'
    files = sorted(glob.glob(f'{datadir}{probe}*.csv'), key=lambda f: dt.datetime.strptime(f[-12:-4], '%Y%m%d'))
    
    for f in files:
        data = pd.read_csv(f, header=0, names=['','T', 'RH', 'P', 'windsp', 'winddir', 'batt', 'tflag',
       'wflag', 'ID'])
        
        data.index = data['']
        data.drop(columns = [''], axis=1, inplace=True)
        
        # flip tflag where RH exceeds 100
        data.loc[data['RH'] > 100.0, 'tflag'] = 1
        
        # save
        fname = f.split('/')[-1]
        data.to_csv(fpath+fname)
```
